
17. Create a new column in the existing dataframe as “'amount_in_usd'” and convert all the non-USD currencies in USD and store them in the same column. 

```python
#currency converting function used from google link, and returns the converted currency if the amount not in USD

def convert_currency(amount, currency):
    if currency == 'USD':
        return amount
    else:
        url = 'https://www.google.com/search?q={} +{} +to +usd '.format(amount, currency)
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        result = soup.find('div', attrs={'class':'BNeawe iBp4i AP7Wnd'}).text
        result = result.split(' ')[0]
        return result
```

```python
df['ORDER_CURRENCY'] = df['ORDER_CURRENCY'].str.replace('HU1', 'HUI')
```

```python
#to store the distinct types of currency

symbols = [x for x in df['ORDER_CURRENCY'].unique()]

dic = {}
for symbol in symbols:
    try:
        dic[symbol] = convert_currency('1', symbol)
    except:
        print(symbol)
```

```python
#lambda function to multiply the conversion rate specified in the symbols and dic{} dictionary
#float to convert the strings to float variable to perform arithmetic operations
#axis=1 to manipulate only rows

df['amount_in_usd'] = df.apply(lambda x: float(x['ORDER_AMOUNT']) * float(dic[x['ORDER_CURRENCY']]), axis=1)
```


18. Check for values “0” in the “'amount_in_usd” column.


19. Create a new column in the existing dataframe “unique_cust_id” by adding 'CUSTOMER_NUMBER' and 'COMPANY_CODE'

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

1. Create a Histogram on DISTRIBUTION_CHANNEL

```python
plt.figure(figsize=(15,5))

# Counting the occurrences of each distribution channel
channel_counts = df['DISTRIBUTION_CHANNEL'].value_counts()

# Creating a bar plot
plt.bar(channel_counts.index, channel_counts.values)

# Set labels and title
plt.xlabel('Distribution Channel')
plt.ylabel('Frequency')

# Adjust x-axis tick labels
plt.xticks(rotation=90)

# Show the bar plot from highest to lowest
plt.show()
```

2. Create a Pie Chart on ORDER_CURRENCY

```python
# Calculate the count of each currency
currency_counts = df['ORDER_CURRENCY'].value_counts()

# Create a pie chart
plt.pie(currency_counts, labels=currency_counts.index, autopct='%1.1f%%')

# Set title
plt.title('Order Currency Distribution')

# Show the pie chart
plt.show()
```

3. Create a line chart PURCHASE_ORDER_TYPE and DISTRIBUTION_CHANNEL

```python
# Group the data by 'PURCHASE_ORDER_TYPE' and 'DISTRIBUTION_CHANNEL' and count the occurrences
grouped_data = df.groupby(['PURCHASE_ORDER_TYPE', 'DISTRIBUTION_CHANNEL']).size().unstack()

# Create a line chart with circle markers
grouped_data.plot(kind='line', marker='|')

# Set labels and title
plt.xlabel('Purchase Order Type')
plt.ylabel('Count')
plt.title('Purchase Order Type by Distribution Channel')

# Removing the legend
plt.legend('', frameon=False)

# Show the line chart
plt.show()
```

4. Create a line plot on ORDER_CREATION_DATE and amount_in_usd

```python
plt.figure(figsize=(10,2))

# Sorting the DataFrame by 'ORDER_CREATION_DATE' column
df.sort_values('ORDER_CREATION_DATE', inplace=True)

# Creating a line plot
plt.plot(df['ORDER_CREATION_DATE'], df['amount_in_usd'])

# Specifying labels and title
plt.xlabel('Order Creation Date')
plt.ylabel('Amount in USD')
plt.title('Amount in USD over Time')

# x-axis labels rotated for better readability
plt.xticks(rotation=90)

# Displaying the line plot
plt.show()
```

5. Create a boxplot on ORDER_AMOUNT	to find out the outliers.

```python
# Converting 'ORDER_AMOUNT' column to numeric
df['ORDER_AMOUNT'] = pd.to_numeric(df['ORDER_AMOUNT'], errors='coerce')

# Creating a horizontal boxplot
plt.boxplot(df['ORDER_AMOUNT'], vert=True)

# Specifying labels and title
plt.xlabel('Amount')
plt.ylabel('Order Amount')
plt.title('Boxplot of Order Amount')

# Displaying the boxplot
plt.show()
```

```python
# Finding outliers
order_amount = df['ORDER_AMOUNT']
q1 = order_amount.quantile(0.25)
q3 = order_amount.quantile(0.75)
iqr = q3 - q1

# Calculation of the upper and lower bounds for outliers
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Identifying outliers
outliers = order_amount[(order_amount < lower_bound) | (order_amount > upper_bound)]
print("Outliers:")
print(outliers)
```

6. Create a barchart on COMPANY_CODE and ORDER_AMOUNT

```python
# Group the data by 'COMPANY_CODE' and calculate the sum of 'ORDER_AMOUNT' for each company code
grouped_data = df.groupby('COMPANY_CODE')['ORDER_AMOUNT'].sum()

# Create a dictionary with the group data
company_amount_dict = dict(grouped_data)
```

```python
plt.figure(figsize=(20,5))

# Extract the keys and values from the dictionary
company_codes = list(company_amount_dict.keys())
order_amounts = list(company_amount_dict.values())

# Create a bar plot
plt.bar(company_codes, order_amounts)

# Set labels and title
plt.xlabel('Company Code')
plt.ylabel('Total Order Amount')
plt.title('Total Order Amount by Company Code')

# Rotate x-axis labels if needed
plt.xticks(rotation=45)

plt.ylim(top=2657627960)

# Show the bar plot
plt.show()
```
