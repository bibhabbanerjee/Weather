```python
#Sanity check - Check how many records are having order date greater than the delivery date
df1.loc[df1.ORDER_CREATION_DATE > df1.REQUESTED_DELIVERY_DATE].shape[0]
```

```python
#Check the number of records where the “ORDER_AMOUNT” field is having “-” in it.
(df1['ORDER_AMOUNT'].str.contains("-")).sum()
```

```python
#Replace “-” with “” from the “ORDER_AMOUNT” field.

df1['ORDER_AMOUNT'] = df1['ORDER_AMOUNT'].str.replace('-', '')
print(df1.ORDER_AMOUNT)
```

```python
#Check the number of records where the “ORDER_AMOUNT” field is having “,” in it..
(df1['ORDER_AMOUNT'].str.contains(",")).sum()
```

```python
#Replace “,” with “.” from the “ORDER_AMOUNT” field.

df1['ORDER_AMOUNT'] = df1['ORDER_AMOUNT'].str.replace(',', '.')
print(df1.ORDER_AMOUNT)
```

```python
#Count the number of records where the order date and the delivery date are same

df1.loc[df1.ORDER_CREATION_DATE == df1.REQUESTED_DELIVERY_DATE].shape[0]
```

```python
#Count the number of records for each currency type by using the field “'ORDER_CURRENCY'”

df1["ORDER_CURRENCY"].value_counts()
```

```python
#  Create a new column in the existing dataframe as “'amount_in_usd'” and convert all the non-USD currencies in USD and store them in the same column. 
cur=df1['ORDER_CURRENCY'].unique()
from currency_converter import CurrencyConverter
from datetime import date
c = CurrencyConverter()
currency_rate={
    'AED':0.27,
    'SAR':0.27,
    'BHD':2.65,
    'KWD':3.25,
    'QAR':0.27
}
for i in cur:
    key=i
    if i=='HU1':
        key='HUF'
    if key in c.currencies:
        currency_rate[i]=c.convert(1, key, 'USD')
def convert(a,currency):
    return currency_rate[currency]*float(a)
df1['AMOUNT_IN_USD']=list(map(convert,df1['ORDER_AMOUNT'],df1['ORDER_CURRENCY']))
df1
```

```python
#Check for values “0” in the “'amount_in_usd” column. 
(df1['AMOUNT_IN_USD']==0).sum()
```

```python
# Create a new column "unique_cust_id" by combining "CUSTOMER_NUMBER" and "COMPANY_CODE"
df1['UNIQUE_CUST_ID'] = df1['CUSTOMER_NUMBER'] + df1['COMPANY_CODE']
print(df.head(10))
```

# MILESTONE 3 - Feature Engineering and Selection

# 1) Check for the outliers in the “amount_in_usd” column and replace the outliers with appropriate values, discussed in the sessions.


```python
import statistics as sts
import seaborn as sns
```

```python
Q1 = np.quantile(df1['AMOUNT_IN_USD'], 0.25)
Q3 = np.quantile(df1['AMOUNT_IN_USD'], 0.75)
IQR = Q3-Q1
low = Q1 - (IQR*1.5)
high = Q3 + (IQR*1.5)

median_value=sts.median(df1['AMOUNT_IN_USD'])

df1['AMOUNT_IN_USD']=np.where((df1.AMOUNT_IN_USD<low)|(df1.AMOUNT_IN_USD>high),median_value,df1['AMOUNT_IN_USD'])
print(df1)
```

# 2)Label encoding or One hot Encoding on all the categorical columns 


```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()

categorial_data=df1.select_dtypes(include='object').columns

for i in range(len(categorial_data)):
    df1[categorial_data[i]]=label_encoder.fit_transform(df1[categorial_data[i]].astype(str))
df1
```

# 3)Log Transformations on continuous columns 

```python
df1.select_dtypes(include=['int64', 'float64']).columns
df1['CUSTOMER_ORDER_ID'] = (df1['CUSTOMER_ORDER_ID']+1).transform(np.log10)
df1['SALES_ORG'] = (df1['SALES_ORG']+1).transform(np.log10)
df1['COMPANY_CODE'] = (df1['COMPANY_CODE']+1).transform(np.log10)
df1['ORDER_CREATION_TIME'] = (df1['ORDER_CREATION_TIME']+1).transform(np.log10)
df1['SOLD_TO_PARTY'] = (df1['SOLD_TO_PARTY']+1).transform(np.log10)
df1['ORDER_AMOUNT'] = (df1['ORDER_AMOUNT']+1).transform(np.log10)
df1['CUSTOMER_NUMBER'] = (df1['CUSTOMER_NUMBER']+1).transform(np.log10)
df1['AMOUNT_IN_USD'] = (df1['AMOUNT_IN_USD']+1).transform(np.log10)
df1['UNIQUE_CUST_ID'] = (df1['UNIQUE_CUST_ID']+1).transform(np.log10)
df1['Last_1st_day_Sales'] = (df1['Last_1st_day_Sales']+1).transform(np.log10)
```

# 4)Try to extract new features by grouping existing columns 


```python
df1['amount_per_customer'] = df1.groupby('CUSTOMER_NUMBER')['ORDER_AMOUNT'].transform('sum')
df1['amount_per_customer'] = (df1['amount_per_customer']+1).transform(np.log10)
df1
```
