```python
#  Create a new column in the existing dataframe as “'amount_in_usd'” and convert all the non-USD currencies in USD and store them in the same column. 
cur=df1['ORDER_CURRENCY'].unique()
from currency_converter import CurrencyConverter
from datetime import date
c = CurrencyConverter()
currency_rate={
    'AED':0.27,
    'SAR':0.27,
    'BHD':2.65,
    'KWD':3.25,
    'QAR':0.27
}
for i in cur:
    key=i
    if i=='HU1':
        key='HUF'
    if key in c.currencies:
        currency_rate[i]=c.convert(1, key, 'USD')
def convert(a,currency):
    return currency_rate[currency]*float(a)
df1['AMOUNT_IN_USD']=list(map(convert,df1['ORDER_AMOUNT'],df1['ORDER_CURRENCY']))
df1
```

# MILESTONE 3 - Feature Engineering and Selection

# 1) Check for the outliers in the “amount_in_usd” column and replace the outliers with appropriate values, discussed in the sessions.


```python
import statistics as sts
import seaborn as sns
```

```python
import numpy as np
import statistics as sts

# Calculate the 1st quartile (Q1), 3rd quartile (Q3), and interquartile range (IQR)
Q1 = np.quantile(df1['AMOUNT_IN_USD'], 0.25)
Q3 = np.quantile(df1['AMOUNT_IN_USD'], 0.75)
IQR = Q3 - Q1

# Calculate the lower and upper bounds for outlier detection
low = Q1 - (IQR * 1.5)
high = Q3 + (IQR * 1.5)

# Calculate the median value of 'AMOUNT_IN_USD'
median_value = sts.median(df1['AMOUNT_IN_USD'])

# Replace the outliers with the median value
df1['AMOUNT_IN_USD'] = np.where((df1['AMOUNT_IN_USD'] < low) | (df1['AMOUNT_IN_USD'] > high), median_value, df1['AMOUNT_IN_USD'])

# Print the modified DataFrame
print(df1)
```

# 6)Perform Hyperparameter tuning, select best hyperparameters by using appropriate algorithms

# XgBoost

```python
from sklearn.model_selection import RandomizedSearchCV

# Define the parameter grid
param_grid = {
    'n_estimators': np.arange(50, 300, 10),
    'max_depth': np.arange(3, 15),
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'subsample': [0.6, 0.7, 0.8, 0.9, 1.0],
    'colsample_bytree': [0.6, 0.7, 0.8, 0.9, 1.0]
}

# Create an XGBoost regressor model
xgb_model = xgb.XGBRegressor()

# Perform random search with cross-validation
random_search = RandomizedSearchCV(estimator=xgb_model, param_distributions=param_grid, n_iter=10,
                                   scoring='r2', cv=5, verbose=1, random_state=42, error_score='raise')

# Fit the random search on the training data
random_search.fit(X_train, y_train)

# Get the best parameters and model
best_param = random_search.best_params_
best_model = random_search.best_estimator_

# Make predictions on the test set using the best model
y_pred = best_model.predict(X_test)

# Calculate evaluation metrics
best_mse = mean_squared_error(y_test, y_pred)
best_rmse = best_mse ** 0.5
best_r2 = r2_score(y_test, y_pred)

# Print the best parameters and scores
print("Best parameters found: ", best_param)
print("Best R2 score: ", random_search.best_score_)
print("After using best Parameters:-")
print("MSE:", best_mse)
print("RMSE:", best_rmse)
print("R2 Score:", best_r2)
```

##### So after performing the HyperParameter tuning on XgBoost model I get the accuracy of the XgBoost model is more so I decided to use XgBoost for this project
