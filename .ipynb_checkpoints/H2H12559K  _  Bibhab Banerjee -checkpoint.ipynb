```python
# 6. Replace all the null values with "NaN"
df.fillna("NaN", inplace=True)
df.isnull()
```

```python
# 7. Change the format of date columns - "ORDER_CREATION_DATE" to datetime[64] with the format as "%Y%m%d"
df['ORDER_CREATION_DATE'] = pd.to_datetime(df['ORDER_CREATION_DATE'], format="%Y%m%d")
df.describe()
```

```python
# 8. Do the same activity for the other date field i.e. "REQUESTED_DELIVERY_DATE" to datetime[64] with the format as "%Y%m%d"
df['REQUESTED_DELIVERY_DATE'] = pd.to_datetime(df['REQUESTED_DELIVERY_DATE'], format="%Y%m%d")
df.describe()
```

```python
# 9. Sanity check - Check how many records are having order date greater than the delivery date
countcheck = len(df[df['ORDER_CREATION_DATE'] > df['REQUESTED_DELIVERY_DATE']])
print('Total count is : ', countcheck)
```

```python
# 10. Remove those records where order date is greater than the delivery date
deletingIndex = df[(df['REQUESTED_DELIVERY_DATE'] < df['ORDER_CREATION_DATE'])].index
df = df.drop(deletingIndex)
df
```

```python
# 11. Check the number of records where the “ORDER_AMOUNT” field is having “-” in it.

check = df.loc[df["ORDER_AMOUNT"].str.contains("-")].copy()
print(len(check))
```

```python
# 12. Replace “-” with “” from the “ORDER_AMOUNT” field. 
df['ORDER_AMOUNT'] = df['ORDER_AMOUNT'].str.replace('-', '')
```

```python
# 13. Check the number of records where the “ORDER_AMOUNT” field is having “,” in it.
checkcomma = df[df['ORDER_AMOUNT'].str.contains(',')]
print(len(checkcomma))
```

```python
# 14. Replace “,” with “.” from the “ORDER_AMOUNT” field. 
df['ORDER_AMOUNT'] = df['ORDER_AMOUNT'].str.replace(',', '.')
```

```python
# 15. Count the number of records where the order date and delivery date are the same

countsamedate = df[df['ORDER_CREATION_DATE'].dt.date == df['REQUESTED_DELIVERY_DATE'].dt.date]

print('Number of records are : ', len(countsamedate))
```

```python
# 16. Count the number of records for each currency type by using the field “'ORDER_CURRENCY'”

countcurrency = df['ORDER_CURRENCY'].value_counts()
print(countcurrency)
```

```python
# 17. Create a new column in the existing dataframe as “'amount_in_usd'” and convert all the non-USD currencies in USD and store them in the same column. 
conversion_rates = {
    'USD': 1,
    'EUR': 1.18,
    'AUD': 0.80,
    'CAD': 0.82,
    'GBP': 1.39,
    'MYR': 0.24,
    'PLN': 0.26,
    'AED': 0.27,
    'HKD': 0.13,
    'CHF': 1.09,
    'RON': 0.24,
    'SGD': 0.75,
    'CZK': 0.046,
    'HU1': 0.0036,
    'NZD': 0.71,
    'BHD': 2.65,
    'SAR': 0.27,
    'QAR': 0.27,
    'KWD': 3.33,
    'SEK': 0.11
}

df['ORDER_AMOUNT'] = pd.to_numeric(df['ORDER_AMOUNT'], errors='coerce')

def calculate_amount_in_usd(row):
    if row['ORDER_CURRENCY'] != 'USD':
        return row['ORDER_AMOUNT'] / conversion_rates.get(row['ORDER_CURRENCY'], 1)
    else:
        return row['ORDER_AMOUNT']

df['amount_in_usd'] = df.apply(calculate_amount_in_usd, axis=1)
```

```python
# 18. Check for values “0” in the “'amount_in_usd” column.
# countzeros = df[df['amount_in_usd'] == 0]
# print(len(countzeros))

zero_amounts = df[df['amount_in_usd'] == 0]

# Display the rows where the amount_in_usd is 0
print(zero_amounts)
```

```python
# 19. Create a new column in the existing dataframe “unique_cust_id” by adding 'CUSTOMER_NUMBER' and 'COMPANY_CODE'
df['unique_cust_id'] = df['CUSTOMER_NUMBER'].astype(str) + '_' + df['COMPANY_CODE'].astype(str)
print(df)
```

```python
# 1.Check for the outliers in the “amount_in_usd” column and replace the
#outliers with appropriate values, discussed in the sessions.

import numpy as np

q1 = np.percentile(df['amount_in_usd'],25)
q3 = np.percentile(df['amount_in_usd'],75)

iqr = q3-q1

lower = q1-1.5*iqr
upper = q3+1.5*iqr

outliers = df[(df['amount_in_usd'] < lower) | (df['amount_in_usd'] > upper)]

replacement = df['amount_in_usd'].median()

df.loc[outliers.index,"amount_in_usd"] = replacement
```
