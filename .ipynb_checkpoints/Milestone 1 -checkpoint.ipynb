```python
import pandas as pd

# Read the CSV file and create a dataframe
df = pd.read_csv('Final.csv')

# Perform operations on the dataframe as needed
# For example, you can access columns using df['column_name']
# or perform calculations and transformations on the data

# Print the first few rows of the dataframe
print(df.head())
```

```python
import pandas as pd

# Assuming the dataframe is named 'df'

# Check the description of the dataframe
description = df.describe()
print(description)

# Check the shape of the dataframe
shape = df.shape
print(shape)

# Check the dataframe information
info = df.info()
print(info)

# Check for null values in the dataframe
null_values = df.isnull().sum()
print(null_values)
```

```python
import pandas as pd

# Assuming the dataframe is named 'df'

# Replace null values with "NaN"
df = df.fillna("NaN")

# Verify the dataframe after replacing null values
print(df)
```

```python
import pandas as pd

# Assuming the dataframe is named 'df'

# Convert "ORDER_CREATION_DATE" to datetime format
df["ORDER_CREATION_DATE"] = pd.to_datetime(df["ORDER_CREATION_DATE"], format="%Y%m%d")

# Verify the dataframe after changing the format
print(df["ORDER_CREATION_DATE"])
```

```python
import pandas as pd

# Assuming the dataframe is named 'df' and the order date column is 'ORDER_CREATION_DATE'
# and the delivery date column is 'DELIVERY_DATE'

# Convert 'ORDER_CREATION_DATE' and 'DELIVERY_DATE' to datetime format if not already done
df['ORDER_CREATION_DATE'] = pd.to_datetime(df['ORDER_CREATION_DATE'], format='%Y%m%d')
df['REQUESTED_DELIVERY_DATE'] = pd.to_datetime(df['REQUESTED_DELIVERY_DATE'], format='%Y%m%d')

# Filter the records where order date is greater than delivery date
records_with_order_gt_delivery = df[df['ORDER_CREATION_DATE'] > df['REQUESTED_DELIVERY_DATE']]

# Get the count of such records
count_order_gt_delivery = len(records_with_order_gt_delivery)

# Print the count
print("Number of records with order date greater than delivery date:", count_order_gt_delivery)
```

```python
import pandas as pd

# Assuming the dataframe is named 'df' and the order date column is 'ORDER_CREATION_DATE'
# and the delivery date column is 'DELIVERY_DATE'

# Convert 'ORDER_CREATION_DATE' and 'DELIVERY_DATE' to datetime format if not already done
df['ORDER_CREATION_DATE'] = pd.to_datetime(df['ORDER_CREATION_DATE'], format='%Y%m%d')
df['REQUESTED_DELIVERY_DATE'] = pd.to_datetime(df['REQUESTED_DELIVERY_DATE'], format='%Y%m%d')

# Remove the records where order date is greater than delivery date
df = df[df['ORDER_CREATION_DATE'] <= df['REQUESTED_DELIVERY_DATE']]

# Verify the dataframe after removing the records
print(df)
```

```python
import pandas as pd

# Assuming the dataframe is named 'df' and the field is 'ORDER_AMOUNT'

# Check the number of records with "-" in the "ORDER_AMOUNT" field
count_dash = len(df[df['ORDER_AMOUNT'] == "-"])
print("Number of records with '-' in ORDER_AMOUNT:", count_dash)

# Replace "-" with an empty string in the "ORDER_AMOUNT" field
df['ORDER_AMOUNT'] = df['ORDER_AMOUNT'].replace("-", "")

# Verify the dataframe after replacing "-"
print(df)
```

```python
import pandas as pd

# Assuming the dataframe is named 'df' and the field is 'ORDER_AMOUNT'

# Check the number of records with "," in the "ORDER_AMOUNT" field
count_comma = len(df[df['ORDER_AMOUNT'].str.contains(",", na=False)])
print("Number of records with ',' in ORDER_AMOUNT:", count_comma)
```

```python
import pandas as pd

# Assuming the dataframe is named 'df' and the field is 'ORDER_AMOUNT'

# Replace "," with "." in the "ORDER_AMOUNT" field
df['ORDER_AMOUNT'] = df['ORDER_AMOUNT'].str.replace(",", ".")

# Verify the dataframe after replacing ","
print(df)
```

```python
import pandas as pd

# Assuming the dataframe is named 'df' and the order date column is 'ORDER_CREATION_DATE'
# and the delivery date column is 'DELIVERY_DATE'

# Convert 'ORDER_CREATION_DATE' and 'DELIVERY_DATE' to datetime format if not already done
df['ORDER_CREATION_DATE'] = pd.to_datetime(df['ORDER_CREATION_DATE'], format='%Y%m%d')
df['REQUESTED_DELIVERY_DATE'] = pd.to_datetime(df['REQUESTED_DELIVERY_DATE'], format='%Y%m%d')

# Count the number of records where the order date and delivery date are the same
count_same_dates = (df['ORDER_CREATION_DATE'].dt.date == df['REQUESTED_DELIVERY_DATE'].dt.date).sum()

# Print the count
print("Number of records with the same order date and delivery date:", count_same_dates)
```

```python
import pandas as pd
from forex_python.converter import CurrencyRates

# Assuming the dataframe is named 'df' and the currency column is 'ORDER_CURRENCY'
# and the amount column is 'ORDER_AMOUNT'

# Create an instance of the CurrencyRates class
c = CurrencyRates()

# Define the base currency as USD
base_currency = 'USD'

# Create a function to convert a given amount from a currency to USD
def convert_to_usd(amount, currency):
    return c.convert(currency, base_currency, amount)

# Create the 'amount_in_usd' column and convert the amounts to USD
df['amount_in_usd'] = df.apply(lambda row: convert_to_usd(row['ORDER_AMOUNT'], row['ORDER_CURRENCY']), axis=1)

# Print the updated dataframe
print(df)
```
